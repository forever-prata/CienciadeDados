```python
# Import pandas e o matplotlib
import pandas as pd
import matplotlib.pyplot as plt

# Importe o dataset autos para um Pandas Dataframe
df = pd.read_csv("autos.csv")
```

```python
# 1. Exibir as 10 primeiras linhas da base
df.head(10)
```

```python
# 2. Verificar o tipo de dados de cada coluna
df.info()
```

```python
# 3. Quantas linhas e colunas há na base?
df.info()
#201 (0 - 200)
```

```python
# 5. Mostrar estatísticas descritivas somente para colunas numéricas
df.select_dtypes(include=['number']).describe()
```

```python
# 6. Listar os diferentes tipos de carroceria (`body_style`). Não é somente listar a coluna body_style
unique_body_styles = df['body-style'].unique()
print(unique_body_styles)
```

```python
# 7. Contar quantos carros existem de cada fabricante (`make`)
make_counts = df['make'].value_counts()
print(make_counts)
```

```python
# 8. Calcular a média do preço (`price`)
average_price = df['price'].mean()
print(average_price)
```

```python
# 9. Mostrar a média de preço por tipo de carroceria
# Calcula a média de preço por tipo de carroceria
average_price_by_body = df.groupby('body-style')['price'].mean()
print(average_price_by_body)
```

```python
# 10. Exibir o carro mais caro da base
most_expensive_car = df.loc[df['price'].idxmax()]
print(most_expensive_car)
```

```python
# 11. Exibir todos os carros com mais de 5 cilindros
cars_over_5_cylinders = df[df['num-of-cylinders'] > 5]
print(cars_over_5_cylinders)
```

```python
# 12. Quantos carros são conversíveis?
num_convertibles = df[df['body-style'] == 'convertible'].shape[0]
print(num_convertibles)
```

```python
# 13. Qual é a média de consumo (city_mpg) dos carros com motor `medium`?
average_city_mpg_medium = df[df['engine_size'] == 'medium']['city-mpg'].mean()
print(average_city_mpg_medium)
```

```python
# 14. Exibir todos os carros com preço acima da média geral
average_price = df['price'].mean()

cars_above_average_price = df[df['price'] > average_price]
print(cars_above_average_price)
```

```python
# 15. Fazer um gráfico de barras com a quantidade de carros por tipo de carroceria

body_style_counts = df['body-style'].value_counts()

plt.figure(figsize=(10, 6))
body_style_counts.plot(kind='bar', color='skyblue')

plt.title('Quantidade de Carros por Tipo de Carroceria', fontsize=14)
plt.xlabel('Tipo de Carroceria', fontsize=12)
plt.ylabel('Quantidade de Carros', fontsize=12)

plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
